# Part 7 — Notes and scope

## Design decisions

**Two aspects.** Separate `mechanism` and `complex` aspects make every protein one
node-layer and remove the identity coupling a flattened model needs. Two
trade-offs follow. The aspects do not share a node population; the translation and
TF-anchor coupling bridge them. The signaling network splits by assembly state, so
code that needs the whole signaling network rejoins it (06 does this inside igraph
on `edge_kind`). A single-`mechanism` model, with complexes left as hyperedges,
avoids both and remains a valid alternative.

**Attribution.** Vertex `kind` drives the heterogeneous export. `edge_kind` drives
the GNN channels. Native hyperedges carry the signed stoichiometry and complex
membership. Slices carry the compartments. The aspect model adds the product
structure and the coupling.

## Results

The Part 4 queries are the clearest payoff of the single container. The Q1 join
spans the regulatory, coupling, and complex layers and returns 1,290 (TF, complex)
pairs, against a mean of 0.10 under 50 random permutations.

The Part 6 GNN reports no accuracy gain. The `co_complex` and `co_metabolic`
projections make the complex and metabolic layers visible to the model. The
ablation then raised held-out AUROC by 0.033 (0.732 → 0.765), within the seed
standard deviation, so the task is regulatory-driven. AnnNet's contribution is
that six sources fuse into one typed graph and ablate layer-by-layer in a handful
of cells.

## Scope

This is a software case study. Fixed seeds, cached downloads, and deterministic
ordering make the build reproducible. The GNN section skips when the deep-learning
stack is absent. No result claims that a predicted edge is a real regulatory
interaction, or that a query output is a biological discovery.
