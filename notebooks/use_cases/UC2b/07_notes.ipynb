# Part 7 - Notes and scope

## What UC2b changes relative to UC2

UC2b keeps UC2's heterogeneous, multi-source build but swaps the **second Kivela
aspect** from *structure* (`complex` = member / monomer) to *time*. Both aspects
now carry biological meaning: `mechanism` says *which process*, `time` says *when
an entity responds*. The change ripples through the series but leaves the AnnNet
machinery - aspects, hyperedges, slices, coupling, history, the four exporters -
untouched.

Three consequences worth naming.

**A real gate replaces a near-null one.** UC2 gated its universe on a generic
protein-atlas evidence flag that retained almost everything. UC2b gates on
*measured response in this cell system*: a gene enters a time layer only where the
kidney-fibrosis study calls it differential there (adj.P < 0.05, symbol union
across four modalities). The universe is now context-specific by construction.

**Coupling becomes earned, not asserted.** UC2's `trans:` edges were a modelling
fiat. UC2b adds **time coupling** - diagonal edges linking the same entity across
consecutive responsive layers - which is a genuine interlayer structure the static
design could not express. The `05` invariant confirms all 7,660 are consecutive and
same-entity.

**The member/monomer split disappears.** UC2's `07` apologised that the `complex`
aspect split the signaling network by assembly state. With `time` as the second
aspect, complexes are simply hyperedges on the `0h` baseline layer, and the
signaling network is whole within each timepoint.

## Design decisions

**Pure 0-hop responsive restriction.** A node-layer `(v, mechanism, t)` exists only
where `v` is responsive at `t`; an intra-layer edge survives only if both endpoints
are. A sizing analysis rejected the alternatives: seed + 1-hop expansion against the
prior-knowledge network over-densified (consecutive layers 64-92% identical, so the
time aspect carried no structure), while a fold-change floor starved the early
layers. Pure restriction on the symbol union gives layers that genuinely differ and
grow with the response (137 -> 10,920 signaling edges, 1h -> 96h).

**Attribute placement.** AnnNet keeps one attribute record per `vertex_id`, so
per-timepoint signal cannot live on the vertex. Time-resolved values stay in
`responsive.parquet`; the graph carries the time signal in *layer membership* and
*edge coordinates*, which is where the queries and the GNN read it.

## Results

**Queries (04).** The temporal join is the clearest payoff. The Q1 co-regulation
program grows from 3 (TF, complex) pairs at 1h to 278 at 96h; cross-compartment
signaling rises from 85 to 6,565 edges. Each is the same few-line query evaluated
per layer.

**Validation (05).** Construction invariants hold exactly (0 responsive
violations). Against the study's *independent* CARNIVAL solution, the raw
responsive graph already contains a growing fraction of the solved network
(early 6% -> late 24%) with no optimisation of its own - a modest, directional
signal. The overlap is limited partly because UC2b uses OmniPath + DoRothEA while
CARNIVAL solved over a richer merged prior-knowledge network; the comparison is a
sanity check, not a benchmark.

**Forecasting (06).** Given only the <= 72h multilayer graph, a heterogeneous
GraphSAGE forecasts the 96h-specific responders (7% base rate) at AUROC ~0.93. As
in UC2, the multi-source channels move the metric only within the seed spread
(std ~0.01); the time-coupling channel raises average precision (0.30 -> 0.38). The
result is secondary. The point is that a temporal, multilayer, multi-source
forecasting task - node index, six edge families, layer ablation - is a handful of
cells on one object.

## Scope

This is a software case study. Fixed seeds, cached downloads, and deterministic
ordering make the build reproducible; the GNN section skips when the deep-learning
stack is absent. No result claims that a forecast gene is a real regulator, that a
query output is a discovery, or that any finding about kidney fibrosis follows. The
biology is a substrate for exercising one heterogeneous, time-resolved container.
