# Part 8 — Notes and scope

This series is an AnnNet software case study. It shows one AnnNet object carrying,
in a single container, everything a multi-condition causal-signaling analysis
needs:

- the OmniPath prior-knowledge network, with signed/directional edge metadata and
  annotation sources (`01`);
- per-patient multi-omics state as vertex-layer attributes — no parallel
  dict-of-dicts (`02`–`03`);
- per-patient CARNIVAL causal subnetworks written back as intra-layer edges, and a
  `carnival_inferred` slice that overlays the inferred topology on the prior
  without a rebuild (`04`);
- a cohort consensus layer plus the inter-layer coupling that makes the patient
  layers genuinely Kivelä-multilayer (`05`);
- degree-bias, null-cohort, and pathway sanity checks (`06`);
- multi-backend access (igraph PageRank written back as a layer attribute), PyG
  and CX2 export, and a provenance diff reconstructed from stage snapshots (`07`).

**Scope.** Fixed seeds, cached downloads, and recorded solver status make the
build reproducible. Nothing here claims that an inferred edge is a real regulatory
interaction or that a consensus hub is a biological discovery; external validation
and method-vs-method benchmarks are out of scope.
