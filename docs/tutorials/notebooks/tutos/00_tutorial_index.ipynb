# AnnNet tutorials

A short, progressive tour. Each lesson is self-contained and assumes only the previous ones.

| # | Lesson | What you learn |
|---|---|---|
| 01 | [Quickstart](01_quickstart.ipynb) | Create a graph, add vertices and edges, inspect it. 5 minutes. |
| 02 | [Attributes & views](02_attributes_and_views.ipynb) | Attach metadata to vertices/edges; read it back through views. |
| 03 | [IO & interop](03_io_and_interop.ipynb) | Save / load (native, parquet, JSON), bridge to NetworkX and igraph. |
| 04 | [Slices](04_slices.ipynb) | Condition / cohort subsets, slice algebra, per-slice weight overrides. |
| 05 | [Hyperedges](05_hyperedges.ipynb) | n-ary edges: undirected complexes and directed reactions. |
| 06 | [Multilayer](06_multilayer.ipynb) | Aspects, supra-nodes, intra/inter/coupling edges. |

For deeper dives, see the **Special Topics** notebooks in the sidebar — directed hyperedges & stoichiometry, multilayer math, and flexible edge orientation.
